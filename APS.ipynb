```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; 
```

```python
estado_expec=data.loc[:,["ano","sigla_uf","expectativa_vida"]]
ano_1991=estado_expec.query("ano==1991")
ano_1991
```
